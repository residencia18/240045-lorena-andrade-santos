**Exercício 1: Utilizando estruturas de controle de fluxo e listas.**

```python
tarefas = []
id = 1;
while True:
    print("1 - Listar Tarefas")
    print("2 - Registrar Tarefa")
    print("3 - Marcar Tarefa como Realizada")
    print("4 - Editar Tarefa")
    print("0 - Sair")
    
    opcao = input("Escolha uma opção: ")
    
    if opcao == "1":
        print("Tarefas registradas:")
        print("-" * 20)
        for tarefa in tarefas:
            print(f"{tarefa['id']}. {tarefa['descricao']} {'[x]' if tarefa['realizada'] else '[ ]'}")
        print("-" * 20)
    elif opcao == "2":
        descricao = input("Digite a descrição da tarefa: ")
        if descricao == "":
            print("Não é possível criar uma tarefa com descricao em branco")
        else:
            tarefa = {'id': id, 'descricao': descricao, 'realizada': False}
            tarefas.append(tarefa)
            id=id+1;
        print("-" * 20)
    elif opcao == "3":
        encontrou = False
        print("Marcar uma tarefa como realizada:")
        id = int(input("Digite o Id da tarefa: "))
        for tarefa in tarefas:
            if tarefa['id'] == id:
                encontrou = True
                if not tarefa['realizada']:
                    tarefa['realizada'] = True
                    print(f"Tarefa: {tarefa['descricao']} - marcada como concluída.")
                break
        if not encontrou:
            print(f"Tarefa com o ID:{id} não encontrada.")
        print("-" * 20)
    elif opcao == "4":
        encontrou = False
        print("Editar uma tarefa:")
        id = int(input("Digite o Id da tarefa: "))
        for tarefa in tarefas:
            if tarefa['id'] == id:
                print(f"Tarefa atual: {tarefa['descricao']}.")
                descricao = input("Digite a nova descrição da tarefa: ")
                tarefa['descricao'] = descricao
                print(f"Tarefa: {tarefa['descricao']} - alterada com sucesso.")
                encontrou = True
                break
        if not encontrou:
            print(f"Tarefa com o ID:{id} não encontrada.")
        print("-" * 20)
    elif opcao == "0":
        print("Saindo da lista de tarefas...")
        break
    else:
        print("Opção inválida. Tente novamente.")
        print("-" * 20)
```

**Exercício 2: Pesquisa sobre persistência de dados.**

Existem várias maneiras de realizar a persistência de dados em Python, algumas das mais comuns são:
1. Leitura e escrita de arquivos:
    * Pode-se utilizar as funções *open*, *read*, *write* e *close* para manipular arquivos de texto em Python.
2. Formato JSON:
    * O formato JSON (JavaScript Object Notation) é frequentemente usado para estruturas de dados mais complexas. 
    * O módulo json em Python facilita a serialização e desserialização de objetos em JSON.
3. Banco de Dados:
    * Utilizado para projetos mais complexos. Pode-se utilizar um banco de dados, como SQLite, MySQL, PostgreSQL, etc.
